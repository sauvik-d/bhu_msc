```python
import numpy as np
import pandas as pd
```

```python
def SVD(X, k):
    X_mean = np.mean(X, axis=0)
    X_centered = X - X_mean

    U, S, Vt = np.linalg.svd(X_centered, full_matrices=False)

    V_k = Vt[:k].T

    return U, S, Vt
```

```python
X = np.random.rand(100,5)

U, S, Vt = SVD(X, k=2)

```
