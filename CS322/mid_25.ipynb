```python
import numpy as np
import pandas as pd
```

```python
X = np.random.randn(100, 200)

X.shape
```

```python
y = np.random.randint(0, 10, size=(100,))
```

```python
y
```

```python
Y = np.zeros((100, 10))
```

```python
Y.shape
```

```python
Y[np.arange(100), y] = 1
```

```python
Y
```

```python
W1 = np.random.randn(20, 200)
b1 = np.random.randn(20, 1)
W2 = np.random.randn(10, 20)
b2 = np.random.randn(10, 1)
W1.shape
W2.shape

# first 200 input size so 20x200 then 20 neurons to pass through 10 neurons so 10x20
```

```python
def sm(z): # ex / sum(ex0), we have to pass the whole vector through this function so that we can do the summation below
    e_z = np.exp(z - np.max(z))
    return e_z / np.sum(e_z, axis=0, keepdims=True)
```

```python
def fwd(X): # forward propagation
    z1 = np.dot(W1, X.T) + b1
    a1 = np.tanh(z1)
    z2 = np.dot(W2, a1) + b2
    y = sm(z2)
    return y
```

```python
y_hat = fwd(X)
```

```python
m = 100

loss = -np.sum(Y.T*np.log(y_hat + 1e-8))/m

print("Loss = ", loss)
```

```python
y_hat
```

```python
y_pred = np.argmax(y_hat, axis=0)
y_pred
```

```python
y_pred
```

```python
y
```

```python
confusion_matrix = np.zeros((10, 10), dtype=int)

for i in range(len(y)):
    confusion_matrix[y[i]][y_pred[i]] += 1

confusion_matrix
```

```python
acc = np.trace(confusion_matrix) / np.sum(confusion_matrix)
```

```python
acc
```

```python
prec = []
recall = []
f1 = []

for i in range(10):
    tp = confusion_matrix[i][i]
    fp = np.sum(confusion_matrix[:, i]) - tp
    fn = np.sum(confusion_matrix[i, :]) - tp

    p = tp/(tp+fp)
    r = tp/(tp+fn)
    f = 2*p*r / (p+r+1e-8)

    prec.append(p)

    f1.append(f)
    recall.append(r)
```

```python
p_value = np.mean(prec)
r_value = np.mean(recall)
f_value = np.mean(f1)
```
