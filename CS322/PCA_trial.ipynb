```python
import numpy as np
```

```python
def PCA(X, k):
    X_mean = np.mean(X, axis = 0)
    X_centered = X - X_mean

    covm = np.cov(X_centered, rowvar=False)
    eva, evec = np.linalg.eig(covm)

    sort_idx = np.argsort(eva)[::-1]

    evec = evec[:, sort_idx]
    eva = eva[sort_idx]

    pc = evec[:, :k]

    X_reduced = np.dot(X_centered, pc)

    return X_reduced, pc, evec[:k]
```

```python
X = np.random.rand(100, 5)
```

```python
X_reduced, components, eig_vals = PCA(X, 2)
```
