```python
import torch
import torch.nn as nn
import torch.optim as optim

# -----------------------------
# CNN Model
# -----------------------------
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Layer 1
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1)
        
        # Layer 2
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        
        # Pooling
        self.pool = nn.MaxPool2d(2, 2)
        
        # Fully connected
        self.fc = nn.Linear(32 * 8 * 8, 6)  # 6 classes
        
        self.relu = nn.ReLU()

    def forward(self, x):
        # Layer 1
        x = self.pool(self.relu(self.conv1(x)))   # (1,32,32) → (16,16,16)
        
        # Layer 2
        x = self.pool(self.relu(self.conv2(x)))   # → (32,8,8)
        
        # Flatten
        x = torch.flatten(x, 1)
        
        # Fully connected
        x = self.fc(x)
        
        return x  # logits (Softmax handled by loss)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
model = CNN()
optimizer = optim.Adam(model.parameters(), lr=0.005)
```

```python
import torch
import torch.nn.functional as F

# Parameters
num_samples = 6000
num_classes = 6
img_size = 32

# Images: values in [0,1]
X = torch.rand(num_samples, 1, img_size, img_size)

# Labels (integers)
y_int = torch.randint(0, num_classes, (num_samples,))

# One-hot encoding
y_onehot = F.one_hot(y_int, num_classes=num_classes).float()

print("X shape:", X.shape)
print("y_onehot shape:", y_onehot.shape)
```

```python
epochs = 10

for epoch in range(epochs):
    
    # Forward (FULL DATASET → Batch GD)
    outputs = model(X)
    
    # Use integer labels (not one-hot)
    loss = criterion(outputs, y_int)
    
    # Backward
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```
